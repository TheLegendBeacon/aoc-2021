```python
from collections import defaultdict

class Grid():
    def __init__(self, consider_diagonals):
        self._grid = defaultdict(lambda: 0)
        self.consider_diagonals = consider_diagonals
    
    def add_point(self, x, y):
        self._grid[x, y] += 1
    
    def add_line(self, start, end):
        
        if self.consider_diagonals:
            # Slower but considers diagonals
            
            direction = [0, 0]
            if start[0] > end[0]:
                direction[0] -= 1
            elif start[0] < end[0]:
                direction[0] += 1
            
            if start[1] > end[1]:
                direction[1] -= 1
            elif start[1] < end[1]:
                direction[1] += 1

            current_coord = [x for x in start]
            
            while True:
                    self.add_point(*current_coord)
                    current_coord[0] += direction[0]
                    current_coord[1] += direction[1]
                    if list(current_coord) == list(end):
                        break
            self.add_point(*end)
        
        else:
            # Faster but only Vertical/Horizontal
            
            if start[0] > end[0] or start[1] > end[1]:
                start, end = end, start

            if start[0] == end[0]:
                for point in range(start[1], end[1]+1):
                    self.add_point(start[0], point)

            elif start[1] == end[1]:
                for point in range(start[0], end[0]+1):
                    self.add_point(point, start[1])
                    
    def get_point_score(self, x, y):
        return self._grid[x, y]
```

```python
x = Grid(consider_diagonals=True)
x.add_line((6, 2), (2, 6))

print(x._grid)
```

```python
# Test Input
test_parser_input = \
'''
0,9 -> 5,9
8,0 -> 0,8
9,4 -> 3,4
2,2 -> 2,1
7,0 -> 7,4
6,4 -> 2,0
0,9 -> 2,9
3,4 -> 1,4
0,0 -> 8,8
5,5 -> 8,2
'''
```

```python
def parser(input_str):
    lines = input_str.split("\n")
    input_outputs = [row.split("->") for row in lines if row != '']
    input_outputs = [[[int(coordinate) for coordinate in coordinates.split(",")] for coordinates in row] for row in input_outputs]    
    return input_outputs
    
test_inputs = parser(test_parser_input)
test_inputs
```

```python
def get_intersections(inputs, consider_diagonals=False):
    count = 0
    grid = Grid(consider_diagonals)
    for input_coord in inputs:
        grid.add_line(*input_coord)
    for coordinate in grid._grid:
        if grid.get_point_score(*coordinate) > 1:
           count += 1
    return count
```
