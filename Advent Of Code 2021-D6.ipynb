```python
# Bad method; won't work for part 2; contains 1 entry for each fish

def fishtimercalculator(fish_input):
    output = []
    for fish in fish_input:
        if fish > 0:
            output.append(fish-1)
        elif fish == 0:
            output.append(6)
            output.append(8)
    return output
```

```python
from collections import Counter

def fish_after_days(fish_input, day_num):
    x = fishtimercalculator(fish_input)
    for _ in range(day_num-1):
        x = fishtimercalculator(x)
    return x

test_result = fish_after_days([3,4,3,1,2], 18)
test_answer = Counter([6,0,6,4,5,6,0,1,1,2,6,0,1,1,1,2,2,3,3,4,6,7,8,8,8,8])

test_answer == Counter(test_result)
```

```python
from collections import deque

# Faster and better method; the list has number of fishes rather than one entry for each fish

def processor2(counts: deque):
    counts[7] += counts[0]
    counts.rotate(-1)

def fishtimercalculator2(fish_input, days):
    counter = [0]*9
    for x in fish_input:
        counter[x] += 1
    
    counter = deque(counter)
    
    for x in range(days):
        processor2(counter)
    
    return sum(counter)

fishtimercalculator2(inputs, 256)
```
