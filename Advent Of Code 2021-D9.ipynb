```python
from typing import List

class Coordinate():
    def __init__(self, x, y):
        self.x = x
        self.y = y

class Grid():
    def __init__(self, inputs: List[List[int]]) -> None:
        self.matrix = inputs
        self.coords = dict()
        for row in range(len(self.matrix)):
            for item in range(len(self.matrix[row])):
                self.coords[item, row] = self.matrix[row][item]
    
    def get_adjacent(self, point: Coordinate):
        points = []
        if point.y < len(self.matrix)-1:
            points.append(self.coords[point.x, point.y+1]) # top
        if point.x > 0:
            points.append(self.coords[point.x-1, point.y])
        if point.y > 0:
            points.append(self.coords[point.x, point.y-1])
        if point.x < len(self.matrix[0])-1:
            points.append(self.coords[point.x+1,point.y])
        return points

    def get_adjacent_coords(self, point: Coordinate):
        points = []
        if point.y < len(self.matrix)-1:
            points.append((point.x, point.y+1)) # top
        if point.x > 0:
            points.append((point.x-1, point.y))
        if point.y > 0:
            points.append((point.x, point.y-1))
        if point.x < len(self.matrix[0])-1:
            points.append((point.x+1,point.y))
        return points
```

```python
lines = ''

with open("D9data", 'r') as f:
    lines += f.read()

def parser(lines):
    lines = list(map(list, lines.split()))
    lines = [list(map(int, line)) for line in lines]
    return lines

lines = parser(lines)
```

```python
def challenge1(grid: Grid):
    score_counter = 0
    for x in grid.coords:
        adjacent = grid.get_adjacent(Coordinate(*x))
        worked = 1
        for adjacent_item in adjacent:
            if grid.coords[x] < adjacent_item:
                pass
            else:
                worked -= 1
                break
        if worked != 0:
            score_counter += (grid.coords[x] + 1)
    return score_counter

challenge1(Grid([[1, 2, 3, 4], [5, 6, 2, 3], [10, 11, 12, 13]]))
```

```python
challenge1(Grid(lines))
```

```python
def find_sink_spots(grid: Grid):
    sink_spots = []
    for x in grid.coords:
        adjacent = grid.get_adjacent(Coordinate(*x))
        worked = 1
        for adjacent_item in adjacent:
            if grid.coords[x] < adjacent_item:
                pass
            else:
                worked -= 1
                break
        if worked != 0:
            sink_spots.append(x)
    return sink_spots

sink_spots = find_sink_spots(Grid(lines))
sink_spots
```

```python
from copy import deepcopy

disposable_lines = deepcopy(lines)
linegrid = Grid(disposable_lines)

basins = {x: [] for x in sink_spots}

def basin_finder(parent, sink_spot):
    adjacent = linegrid.get_adjacent_coords(Coordinate(*sink_spot))
    for x in adjacent:
        if 0 <= linegrid.coords[x] < 9:
            basins[parent].append(x)
            linegrid.coords[x] = -1
            basin_finder(parent, x)

for x in sink_spots:
    basin_finder(x, x)

basins = {x: basins[x] for x in basins if len(basins[x]) != 0}
values = list(basins.values())
len_values = sorted([len(value) for value in values])
len_values.reverse()
final_vals = len_values[:3]
answer = 1
[answer:=answer*x for x in final_vals]
answer
```
